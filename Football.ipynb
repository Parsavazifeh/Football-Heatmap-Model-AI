```python
import os
import cv2
import numpy as np
import pandas as pd
import time
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader, random_split
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm

#########################################
# 1. Preprocessing Functions
#########################################
def crop_arrow(pil_img: Image.Image, crop_ratio: float = 0.12) -> Image.Image:
    """Crop the top portion (arrow region) from the image."""
    w, h = pil_img.size
    top_crop = int(crop_ratio * h)
    return pil_img.crop((0, top_crop, w, h))

def convert_yellow_to_red(pil_img: Image.Image) -> Image.Image:
    """
    Convert yellow heat spots to red:
      - Convert the image to HSV.
      - For pixels in the yellow range, force the hue to 0 (red).
      - Convert back to RGB.
    """
    bgr_img = np.array(pil_img)[:, :, ::-1]  # PIL RGB -> OpenCV BGR
    hsv = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2HSV)
    lower_yellow = np.array([10, 80, 70])
    upper_yellow = np.array([45, 255, 255])
    yellow_mask = cv2.inRange(hsv, lower_yellow, upper_yellow)
    hsv[:,:,0][yellow_mask > 0] = 0  # Set hue to 0 (red)
    bgr_modified = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    rgb_modified = cv2.cvtColor(bgr_modified, cv2.COLOR_BGR2RGB)
    return Image.fromarray(rgb_modified)

def filter_heat_and_preserve_white_lines(pil_img: Image.Image) -> Image.Image:
    """
    Build a heat mask from red regions (after yellowred conversion),
    preserve white pitch lines, and fill non-heat/non-white areas with pitch-green.
    """
    bgr_img = np.array(pil_img)[:, :, ::-1]
    hsv = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2HSV)

    # Build red heat mask
    lower_red1 = np.array([0, 80, 70])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 80, 70])
    upper_red2 = np.array([180, 255, 255])
    red_mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    red_mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    heat_mask = red_mask1 + red_mask2
    kernel = np.ones((3, 3), np.uint8)
    heat_mask = cv2.morphologyEx(heat_mask, cv2.MORPH_OPEN, kernel, iterations=1)
    heat_mask = cv2.morphologyEx(heat_mask, cv2.MORPH_CLOSE, kernel, iterations=1)

    # Build white lines mask
    lower_white = np.array([0, 0, 200])
    upper_white = np.array([180, 30, 255])
    white_mask = cv2.inRange(hsv, lower_white, upper_white)
    white_mask = cv2.morphologyEx(white_mask, cv2.MORPH_OPEN, kernel, iterations=1)
    white_mask = cv2.morphologyEx(white_mask, cv2.MORPH_CLOSE, kernel, iterations=1)

    # Create result image with pitch-green background
    result_bgr = np.zeros_like(bgr_img, dtype=np.uint8)
    pitch_bgr = (81, 215, 170)  # Pitch-green in BGR
    result_bgr[:, :] = pitch_bgr

    # Preserve white lines
    white_idx = white_mask != 0
    result_bgr[white_idx] = bgr_img[white_idx]

    # Preserve heat areas (red regions)
    heat_idx = heat_mask != 0
    result_bgr[heat_idx] = bgr_img[heat_idx]

    result_rgb = cv2.cvtColor(result_bgr, cv2.COLOR_BGR2RGB)
    return Image.fromarray(result_rgb)

def minimal_preprocessing(pil_img: Image.Image) -> Image.Image:
    """
    Minimal preprocessing pipeline:
      1. Crop arrow.
      2. Convert yellowred.
      3. Filter heat & preserve white pitch lines.
    """
    img = crop_arrow(pil_img, crop_ratio=0.12)
    img = convert_yellow_to_red(img)
    img = filter_heat_and_preserve_white_lines(img)
    return img

#########################################
# 2. Mild Data Augmentation
#########################################
mild_augmentation = transforms.Compose([
    transforms.RandomRotation(5),  # small 5 rotation
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.02),
    transforms.RandomHorizontalFlip(p=0.5)
])

#########################################
# 3. Custom Test Dataset
#########################################
class TestDataset(Dataset):
    def __init__(self, test_dir, transform=None):
        self.test_dir = test_dir
        self.transform = transform
        self.image_files = [os.path.join(test_dir, fname)
                            for fname in os.listdir(test_dir)
                            if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = self.image_files[idx]
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, os.path.basename(img_path)

#########################################
# 4. Custom Collate Function for Test Loader
#########################################
def test_collate_fn(batch):
    images, filenames = zip(*batch)
    images = torch.stack(images, dim=0)
    return images, list(filenames)

#########################################
# 5. Denormalization for Visualization
#########################################
def denormalize(tensor, mean, std):
    mean_tensor = torch.tensor(mean).view(-1, 1, 1)
    std_tensor = torch.tensor(std).view(-1, 1, 1)
    return tensor * std_tensor + mean_tensor

#########################################
# 6. Evaluation Function
#########################################
def evaluate(model, data_loader, device, criterion):
    model.eval()
    running_loss = 0.0
    correct_preds = 0
    total_samples = 0
    with torch.no_grad():
        for inputs, labels in data_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct_preds += torch.sum(preds == labels).item()
            total_samples += inputs.size(0)
    avg_loss = running_loss / total_samples
    accuracy = correct_preds / total_samples
    return avg_loss, accuracy

#########################################
# 7. Main Training & Visualization
#########################################
if __name__ == "__main__":
    # ImageNet normalization stats
    mean = [0.485, 0.456, 0.406]
    std  = [0.229, 0.224, 0.225]

    # Compose the transform: minimal preprocessing, then mild augmentation (50% chance),
    # then resize, ToTensor, and Normalize.
    transform = transforms.Compose([
        transforms.Lambda(minimal_preprocessing),
        transforms.RandomApply([mild_augmentation], p=0.5),
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=mean, std=std)
    ])

    # Load training dataset (assumes folder structure: data/train/<position>/)
    data_dir = "/content/drive/MyDrive/Machine Learning/Football Heatmap/train"
    full_train_dataset = datasets.ImageFolder(root=data_dir, transform=transform)

    # Visualize top 5 preprocessed images per category.
    images_by_category = {cls: [] for cls in full_train_dataset.classes}
    for img, label in full_train_dataset:
        cls_name = full_train_dataset.classes[label]
        if len(images_by_category[cls_name]) < 5:
            images_by_category[cls_name].append(img)
        if all(len(lst) >= 5 for lst in images_by_category.values()):
            break

    num_categories = len(full_train_dataset.classes)
    fig, axes = plt.subplots(num_categories, 5, figsize=(15, 3 * num_categories))
    for i, cls_name in enumerate(full_train_dataset.classes):
        for j in range(5):
            img_tensor = images_by_category[cls_name][j]
            img_denorm = denormalize(img_tensor.clone(), mean, std)
            img_np = img_denorm.permute(1, 2, 0).numpy()
            axes[i, j].imshow(np.clip(img_np, 0, 1))
            axes[i, j].axis('off')
            if j == 0:
                axes[i, j].set_title(cls_name, fontsize=14)
    plt.tight_layout()
    plt.show()

    # Split dataset into training and validation sets (80/20 split)
    val_size = int(0.2 * len(full_train_dataset))
    train_size = len(full_train_dataset) - val_size
    train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=0)

    # Load test dataset using TestDataset and the same transform.
    test_dataset = TestDataset(test_dir="/content/drive/MyDrive/Machine Learning/Football Heatmap/test", transform=transform)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=0, collate_fn=test_collate_fn)

    # Setup device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    #########################################
    # Model Setup using EfficientNet-B0 with Regularization
    #########################################
    model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
    num_classes = len(full_train_dataset.classes)
    in_features = model.classifier[1].in_features  # typically 1280
    model.classifier[1] = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(in_features, num_classes)
    )
    model = model.to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-5)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

    #########################################
    # Training Loop with Evaluation
    #########################################
    num_epochs = 20
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct_preds = 0
        total_samples = 0
        epoch_start = time.time()

        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", unit="batch")
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct_preds += torch.sum(preds == labels).item()
            total_samples += inputs.size(0)
            pbar.set_postfix(loss=f"{loss.item():.4f}")

        epoch_loss = running_loss / total_samples
        epoch_acc = correct_preds / total_samples
        val_loss, val_acc = evaluate(model, val_loader, device, criterion)
        epoch_time = time.time() - epoch_start

        print(f"\nEpoch {epoch+1}/{num_epochs} - Time: {epoch_time:.2f}s "
              f"- Train Loss: {epoch_loss:.4f} - Train Acc: {epoch_acc:.4f} "
              f"- Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.4f}")
        scheduler.step(val_loss)

    # Save trained model (optional)
    torch.save(model.state_dict(), "trained_model_efficientnet_minimal.pth")

    #########################################
    # Prediction on Test Data and Save Submission CSV
    #########################################
    model.eval()
    predictions = []
    filenames = []
    with torch.no_grad():
        for batch in test_loader:
            inputs, fnames = batch
            inputs = inputs.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            predictions.extend(preds.cpu().numpy())
            filenames.extend(fnames)

    # Sort submission by filenames (numerically)
    submission = pd.DataFrame({"filename": filenames, "position": [full_train_dataset.classes[p] for p in predictions]})
    submission["filename"] = submission["filename"].astype(str)
    submission = submission.sort_values(by="filename", key=lambda x: x.str.extract(r'(\d+)', expand=False).astype(int))
    submission.to_csv("submission.csv", index=False)
    print("Submission CSV file saved as submission.csv")
```
